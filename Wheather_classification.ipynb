## KNN ALGORITHM

-it is a non parametric algorithm which does not make any assumptions
-it is one of the simplest and easy to implementable algorithm
-it is also called as lazy algorithm, as it does not learns anything during training phase

## KNN STEPS

-compute the distance between the test point vs all training samples
-(use any distance measure like euclidean,manhattan ,minkowski)
-arrange the distances computed in ascending order
-select the k value 
-vote for majority
-class having higher votes in the outcome
