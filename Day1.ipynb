```python
import numpy as np
import matplotlib.pyplot as plt
A =1
f=1
t= np.arange(0,1.1,0.1)
P=0
PI = np.pi #3.14
print(A,f,t,P)
Y=A*np.sin(2*PI*f*t+P)
plt.plot(t,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,100)
S = np.sin(2*np.pi*3*t)
C = np.cos(2*np.pi*3*t)
D = np.tan(2*np.pi*3*t)
plt.subplot(3,1,1)
plt.plot(t,S,'r')
plt.xlabel("Time Period")
plt.ylabel("Amplitude")
plt.title("Sine Wave")
plt.subplot(3,1,2)
plt.plot(t,C,'k')
plt.title("Cosine Wave")
plt.subplot(3,1,3)
plt.plot(t,D,'g')
plt.title("Tangent Wave")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,1000)
S1 = np.sin(2*np.pi*3*t)
S2 = np.sin(2*np.pi*10*t)
S3 = np.sin(2*np.pi*30*t)
S4 = np.sin(2*np.pi*60*t)
plt.subplot(2,2,1)
plt.plot(t,S1)
plt.subplot(2,2,2)
plt.plot(t,S2)
plt.subplot(2,2,3)
plt.plot(t,S3)
plt.subplot(2,2,4)
plt.plot(t,S4)
plt.show()

S = S1+S2+S3+S4  #signal mixing (addition in time domain)
plt.plot(t,S)
plt.show()

#Converting  to Frequency Domain By Applying FFT
Sfft = np.fft.rfft(S) #Converting Time Domain  to Frequency Domain
Flength = np.fft.rfftfreq(len(S),1/1000)
plt.subplot(2,1,1)
plt.plot(t,S)
plt.title("Time Domain")
plt.subplot(2,1,2)
plt.stem(Flength,np.abs(Sfft))
plt.title("Frequency Domain")
plt.tight_layout()
plt.show()
```
