```python
#building ANN for handwritten mnist data
#import libraries
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Flatten,Activation
from tensorflow.keras import layers
```

```python
#download mnist data
(x_train,y_train),(x_test,y_test)=mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

```

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
#x has images, y has corresponding labels
#now randomly pick an image display the label also

# import random
import matplotlib.pyplot as plt

# i=random.randint(1,60000)
plt.imshow(x_train[10])
plt.title(y_train[10])
plt.show()
```

```python
#reshape the x_train and x_test and also normalize
xmtrain=x_train.reshape(-1,784) #reshape from 2d to 1d
xmtrain=xmtrain/255 #divide by 255 to normalize
xmtest=x_test.reshape(-1,784)
xmtest=xmtest/255


xmtest
```

```python
#preprocessing the y using to_catgorical
ymtrain=keras.utils.to_categorical(y_train)
ymtest=keras.utils.to_categorical(y_test)
```

```python
ymtrain[0:5]
```

```python
y_train[0:5]
```

```python
# build the model architecture
model=Sequential()
model.add(Dense(input_shape=(784,),units=256,activation='relu'))
model.add(Dense(units=10,activation='sigmoid'))
model.summary()
```

```python
#display the architecture
from tensorflow.keras.utils import plot_model
plot_model(model,show_shapes=True)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
#for binary classification loss=binary_crossentropy
#for multiclass it is categoricalcrossentropy
```

```python
#train the model
model.fit(xmtrain,ymtrain,epochs=5)
```

```python
#test the test_data
model.evaluate(xmtest,ymtest)
```

```python
#sample test
plt.imshow(x_test[5])
plt.show()
R=model.predict(xmtest[0:1])
print(R)
print("Prediction",np.argmax(R))
```

```python



#save the model architecture and weights
path='/content/drive/MyDrive/Signal AI/model.h5'
model.save(path)
```

## covolution neural network

```python
#building ANN for handwritten mnist data
#import libraries
import tensorflow as tf
from tensorflow import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Flatten,Activation
from tensorflow.keras.layers import Conv2D,MaxPooling2D
```

```python
#download mnist data
(x_train,y_train),(x_test,y_test)=mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

```

```python
#preprocessing , reshaping not required , normalization is optional
xmtrain=x_train/255
xmtest=x_test/255
```

```python
#define the architecture
model=Sequential()
model.add(Conv2D(8,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(xmtrain,y_train,epochs=5,batch_size=100)
```

```python
model.evaluate(xmtest,y_test)
```
